# Importer les Donnees & Lectur des donnees : 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importation des donnees
data = pd.read_csv("quotes.csv")

# des infos sur data
print("la dimention de dataset est : " , data.shape ,"\n")
print(f"le nbr des lignes : {len(data)}")
print(f"le nbr des cologne : {len(data.columns)} \n")
print("le Type des donnees est : \n ", data.dtypes , "\n")

data.head()
```

# Netoyage les Donnees :

```python
# les lignes avec des valeures manquant :
print(f" - nbr des lignes avec des valeurs Manquant sur chaque cologne : \n{data.isnull().sum()}")

# En Observe que jai 3 lignes avec 3 tags non defni , donc en le suprime
data = data.dropna()
print(f"\n - le nbr des lignes sans les ligne manquant : {len(data)}")

# Verifier si il y a des doublant 
print(f" - les nbrs les lignes doublant : {data.duplicated().sum()}")
```
