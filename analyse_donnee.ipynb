# Importer les Donnees & Lectur des donnees : 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importation des donnees
data = pd.read_csv("quotes.csv")

# des infos sur data
print("la dimention de dataset est : " , data.shape ,"\n")
print(f"le nbr des lignes : {len(data)}")
print(f"le nbr des cologne : {len(data.columns)} \n")
print("le Type des donnees est : \n ", data.dtypes , "\n")

data.head()
```

# Netoyage les Donnees :

```python
# les lignes avec des valeures manquant :
print(f" - nbr des lignes avec des valeurs Manquant sur chaque cologne : \n{data.isnull().sum()}")

# En Observe que jai 3 lignes avec 3 tags non defni , donc en le suprime
data = data.dropna()
print(f"\n - le nbr des lignes sans les ligne manquant : {len(data)}")

# Verifier si il y a des doublant 
print(f" - les nbrs les lignes doublant : {data.duplicated().sum()}")
```

# Graphe pour le Top 10 Auteurs par nbr citations 

```python
author_counts = data['Author'].value_counts().head(10)

plt.barh(author_counts.index, author_counts.values )
plt.xlabel('Nombre de citations')
plt.ylabel('Auteur')
plt.title('Top 10 auteurs par nombre de citations' , fontweight='bold')
plt.gca().invert_yaxis()  # Pour avoir le plus grand en haut

plt.show()
```

# Quelque analyses important : 

```python
print(f" - Nombre total de citations : {len(data)}")
print(f" - Nombre d'auteurs uniques : {data['Author'].nunique()}")
print(f" - Nombre de pages scrapées: {data['Page'].nunique()}")
```

# Repartition des 10 auteurs principaux

```python
plt.figure(figsize=(10, 8))
plt.pie(author_counts.values, labels=author_counts.index, autopct='%1.1f%%')
plt.title('Repartition des 10 auteurs principaux', fontsize=14 , fontweight='bold')
plt.show()
```
